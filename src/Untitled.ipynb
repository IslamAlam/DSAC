```python
import matplotlib.gridspec as gridspec
from matplotlib.backends.backend_pdf import PdfPages


def multipage(filename, figs=None, dpi=300):
    pp = PdfPages(filename)
    if figs is None:
        figs = [plt.figure(n) for n in plt.get_fignums()]
    for fig in figs:
        #fig.savefig(pp, format='pdf')
        pp.savefig(fig) #, format='pdf')
    # pp.close()
    return pp

def save_plot_snakes(snake,snake_hist,GT,mapE, mapA, mapB, mapK, grads_arrayE, grads_arrayA, grads_arrayB, grads_arrayK, image, mask, plot_title, filename):
    
    fig = plt.figure(figsize=(28, 16),frameon=False )# , constrained_layout=True)

    outer = gridspec.GridSpec(2, 3, wspace=0.1, hspace=0.1)
    f_ax = fig.add_subplot(outer[:, :])
    f_ax.axis('off')
    f_ax.set_title(plot_title, fontsize=20)

    inner = [ gridspec.GridSpecFromSubplotSpec(1, 5,
                        subplot_spec=outer[0:1,:], wspace=0.2, hspace=0.2), 
                        gridspec.GridSpecFromSubplotSpec(1, 1,
                        subplot_spec=outer[-1:,0], wspace=0.2, hspace=0.2),
                        gridspec.GridSpecFromSubplotSpec(2, 2,
                        subplot_spec=outer[-1:,1], wspace=0.2, hspace=0.2),
                        gridspec.GridSpecFromSubplotSpec(2, 2,
                        subplot_spec=outer[-1:,2], wspace=0.2, hspace=0.2)]

    # Plot result
    # fig0, (ax) = plt.subplots(ncols=5,nrows=1)
    fig1_ax0 = plt.Subplot(fig, inner[0][:, 0])

    im = fig1_ax0.imshow(scipy.misc.imresize(np.abs(image[:,:,:,0]),mapE[:, :, 0, 0].shape))
    for i in range(0,len(snake_hist),5):
        fig1_ax0.plot(snake_hist[i][:, 1], snake_hist[i][:, 0], '-.', color=[i / len(snake_hist), i / len(snake_hist), 1-i / len(snake_hist)], lw=3)
    if not GT is None:
        fig1_ax0.plot(GT[:, 1], GT[:, 0], '-', color=[0.2, 1, 0.2], lw=3)
    fig1_ax0.plot(snake[:, 1], snake[:, 0], '--', lw=3,color=[1,1,0])
    fig1_ax0.axis('off')
    fig1_ax0.set_title(r'a) image $\mathbf{x}$', y=-0.2)
    plt.colorbar(im, ax=fig1_ax0, fraction=0.046, pad=0.04).remove()
    fig.add_subplot(fig1_ax0)

    #plt.colorbar(im, ax=ax)
    #fig0.suptitle('Image, GT (red) and converged snake (black)', fontsize=20)
    fig1_ax1 = plt.Subplot(fig, inner[0][:, 1])
    im0 = fig1_ax1.imshow(mapE[:, :, 0, 0])
    plt.colorbar(im0, ax=fig1_ax1,fraction=0.046, pad=0.04)
    fig1_ax1.axis('off')
    fig1_ax1.set_title(r'b) data term $D(\mathbf{x})$', y=-0.2)
    fig.add_subplot(fig1_ax1)

    fig1_ax2 = plt.Subplot(fig, inner[0][:, 2])
    im1 = fig1_ax2.imshow(mapK[:, :, 0, 0])
    plt.colorbar(im1, ax=fig1_ax2,fraction=0.046, pad=0.04)
    fig1_ax2.axis('off')
    fig1_ax2.set_title(r'c) balloon term $\kappa(\mathbf{x})$', y=-0.2)
    fig.add_subplot(fig1_ax2)

    fig1_ax3 = plt.Subplot(fig, inner[0][:, 3])
    im2 = fig1_ax3.imshow(mapB[:, :, 0, 0])
    plt.colorbar(im2, ax=fig1_ax3,fraction=0.046, pad=0.04)
    fig1_ax3.axis('off')
    fig1_ax3.set_title(r'd) thin plate term $ \beta(\mathbf{x})$', y=-0.2)
    fig.add_subplot(fig1_ax3)

    fig1_ax4 = plt.Subplot(fig, inner[0][:, 4])
    im3 = fig1_ax4.imshow(mapA[:, :, 0, 0])
    plt.colorbar(im3, ax=fig1_ax4, fraction=0.046, pad=0.04)
    fig1_ax4.axis('off')
    fig1_ax4.set_title(r'd) membrane term $ \alpha(\mathbf{x})$', y=-0.2)
    fig.add_subplot(fig1_ax4)


    if not grads_arrayE is None:
        # fig1, (ax) = plt.subplots(ncols=1)
        fig2_ax0 = plt.Subplot(fig, inner[1][:, :])
        fig.add_subplot(fig2_ax0)
        im = fig2_ax0.imshow(mask[:, :, 0, 0])
        fig2_ax0.plot(GT[:, 1], GT[:, 0], '--r', lw=3)
        plt.colorbar(im, ax=fig2_ax0)

        fig3 = plt.Subplot(fig, inner[2][:, :])
        fig.add_subplot(fig3)
        fig3.axis('off')
        fig3.set_title('Output maps', fontsize=20)

        fig3_ax0 = plt.Subplot(fig, inner[2][0, 0])
        fig.add_subplot(fig3_ax0)

        fig3_ax1 = plt.Subplot(fig, inner[2][0, 1])
        fig.add_subplot(fig3_ax1)

        fig3_ax2 = plt.Subplot(fig, inner[2][1, 0])
        fig.add_subplot(fig3_ax2)

        fig3_ax3 = plt.Subplot(fig, inner[2][1, 1])
        fig.add_subplot(fig3_ax3)

        # fig2, fig3_ax0 = plt.subplots(ncols=2,nrows=2)
        im0 = fig3_ax0.imshow(mapE[:, :, 0, 0])
        plt.colorbar(im0, ax=fig3_ax0)
        fig3_ax0.set_title('D')
        im1 = fig3_ax0.imshow(mapA[:, :, 0, 0])
        plt.colorbar(im1, ax=fig3_ax1)
        fig3_ax1.set_title('alpha')
        im2 = fig3_ax2.imshow(mapB[:, :, 0, 0])
        plt.colorbar(im2, ax=fig3_ax2)
        fig3_ax2.set_title('beta')
        im3 = fig3_ax3.imshow(mapK[:, :, 0, 0])
        plt.colorbar(im3, ax=fig3_ax3)
        fig3_ax3.set_title('kappa')
        # fig3_ax3.title('Output maps', fontsize=20)

        fig4 = plt.Subplot(fig, inner[3][:, :])
        fig.add_subplot(fig4)
        fig4.axis('off')
        fig4.set_title('Gradient maps', fontsize=20)

        fig4_ax0 = plt.Subplot(fig, inner[3][0, 0])
        fig.add_subplot(fig4_ax0)

        fig4_ax1 = plt.Subplot(fig, inner[3][0, 1])
        fig.add_subplot(fig4_ax1)

        fig4_ax2 = plt.Subplot(fig, inner[3][1, 0])
        fig.add_subplot(fig4_ax2)

        fig4_ax3 = plt.Subplot(fig, inner[3][1, 1])
        fig.add_subplot(fig4_ax3)
        # fig3, ax = plt.subplots(ncols=2,nrows=2)
        im0 = fig4_ax0.imshow(grads_arrayE[:, :, 0, 0])
        plt.colorbar(im0, ax=fig4_ax0)
        fig4_ax0.set_title('D')
        im1 = fig4_ax1.imshow(grads_arrayA[:, :, 0, 0])
        plt.colorbar(im1, ax=fig4_ax1)
        fig4_ax1.set_title('alpha')
        im2 = fig4_ax2.imshow(grads_arrayB[:, :, 0, 0])
        plt.colorbar(im2, ax=fig4_ax2)
        fig4_ax2.set_title('beta')
        im3 = fig4_ax3.imshow(grads_arrayK[:, :, 0, 0])
        plt.colorbar(im3, ax=fig4_ax3)
        fig4_ax3.set_title('kappa')
        # fig4_ax3.suptitle('Gradient maps', fontsize=20)

    # plt.show()
    pp = multipage(filename, figs=[fig], dpi=300)
    return fig, pp
```
